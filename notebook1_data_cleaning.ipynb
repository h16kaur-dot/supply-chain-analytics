```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('data/raw/DataCoSupplyChainDataset.csv', encoding='latin-1')
print("Shape:",df.shape)
print("Columns:", df.columns.tolist())
```

```python
df.head()
```

```python
missing = df.isnull().sum()
missing_percent = ((df.isnull().sum()/len(df))*100)  
missing_df = pd.DataFrame({"missing values:": missing, "missing percent:": missing_percent}) 
print(missing_df[missing_df['missing values:'] > 0]) 
```

```python
print(df.dtypes)
```

```python
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'])
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'])
print("Date columns fixed!")
print(df[['order date (DateOrders)', 'shipping date (DateOrders)']].dtypes)
```

```python
cols_to_drop = ['Customer Password', 'Product Description', 'Product Image']
df = df.drop(columns=cols_to_drop)
print("Remaining columns:", df.shape[1])
```

```python
print("before_dup" ,df.shape)
df=df.drop_duplicates()
print("after_dup" ,df.shape)
```

```python
df['delivery_delay_days'] = (
    df['shipping date (DateOrders)'] - df['order date (DateOrders)']
).dt.days
print(df['delivery_delay_days'].describe())
```

```python
df.to_csv('data/cleaned/supply_chain_cleaned.csv', index=False)
print("Cleaned data saved successfully!")
print("Final shape:", df.shape)
```
